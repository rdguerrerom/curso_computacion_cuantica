Create a quantum circuit that generates a random bit string array.  The function takes two arguments: `bit_len`, which specifies the number of bits used to represent the random numbers, and `shots`, which specifies the number of random numbers that will be generated. `bit_len` has a default value of 8 (the number of bytes that an ASCII character is composed of) and `shots` has a default value of 100. These random numbers are generated taking advantage of quantum fluctuations of vacuum due to at the moment that we use quantum processes in non-ideal situations, those quantum fluctuations come in handy to perfectly randomize data.

The function uses a quantum circuit with a quantum register and a classical register of size `bit_len`  to create the random numbers. Next, the code creates a `QuantumRegister` with `char_len` qubits and a `ClassicalRegister` with `char_len` classical bits. These registers will be used as the input and output for the quantum circuit.

The code then creates a `QuantumCircuit` object and passes the quantum and classical registers as arguments. This creates a quantum circuit that operates on the specified registers.

The code applies a **Hadamard** gate to each qubit in the quantum register. This creates an equal superposition of all possible bit strings of length `bit_len`, which means that the qubits will be in a state where they are simultaneously 0 and 1 (a superposition of all possible states).

The code then measures each qubit in the quantum register and stores the measurement results in the classical register using the `circuit.measure` method. This collapses the state of the qubits, causing them to take on a definite value of 0 or 1.

The code then gets a backend for the IBM Quantum `ibmq_qasm_simulator` using the `provider.get_backend` method. A backend is an object that represents a quantum computer or simulator, and it can be used to run quantum circuits on the corresponding device. The `ibmq_qasm_simulator` is a quantum simulator provided by IBM Quantum, which can be used to simulate the execution of quantum circuits on a quantum computer.

The code then runs the quantum circuit on the `ibmq_qasm_simulator` using the `execute` function. The `execute` function takes the quantum circuit and the backend as arguments, and it returns a job object that represents the quantum computation. The shots argument is set to 100 by default, which specifies that the circuit should be run 100 times and the measurement results should be averaged to get a more accurate estimate of the probabilities of the different measurement outcomes.

The code then uses the `job_monitor` function to monitor the progress of the quantum computation. This function prints information about the status of the job to the terminal, so the user can see how the computation is progressing.


Then, the code retrieves the measurement results from the `job` object using the `job.result().get_counts` method. This returns a dictionary that maps each bit string to the number of times it was observed as the measurement outcome. The `experiment=None` argument is used to specify that the results from all experiments should be combined. Next, the function filters the measurement results to only keep byte strings that were measured an odd number of times with a probability of 0.5, and even number of times with a probability of 0.5. This is achieved by using a second quantum circuit to generate a random integer that is used to determine whether to keep bit strings that were measured an odd or even number of times. The second quantum circuit contains two qubits, and applies a Hadamard gate to each qubit. The qubits are then measured, and the measurement results are used to generate a random integer in the range 0-1. This integer is then used to determine whether to keep bit strings that were measured an odd or even number of times.

This effectively randomizes the order of the measurement counts in the list.
Finally, the function returns the list of measurement counts.

```python
def magic_randomizer(bit_len=8, shots=100):
    # Circuit for byte strings
    q1 = QuantumRegister(bit_len,'q')
    c1 = ClassicalRegister(bit_len,'c')
    circuit = QuantumCircuit(q1, c1)
    circuit.h(q1)
    circuit.measure(q1,c1)

    # Circuit for random choice of strings
    q2 = QuantumRegister(1,'q')
    c2 = ClassicalRegister(1,'c')
    circuit2 = QuantumCircuit(q2, c2)
    circuit2.h(q2)
    circuit2.measure(q2,c2)

    backend = provider.get_backend('ibmq_qasm_simulator')

    # Execute circuit 2
    job = execute(circuit2, backend, shots=1)
    job_monitor(job)
    randint =  int(list(job.result().get_counts(experiment=None))[0], 2)

    # Execute circuit 1
    job = execute(circuit, backend, shots=shots)
    job_monitor(job)
    counts_res = [k for k, v in job.result().get_counts(experiment=None).items() if v%2 != randint]
    return counts_res
```

## Secure password generator

It is generally considered that longer passwords are more secure than shorter ones. This is because longer passwords have a higher entropy, which means that they have a larger number of possible combinations and are therefore harder to guess or crack.

In general, passwords with 12 or more characters are considered to be sufficiently long and secure for most purposes. However, the actual security of a password depends on a number of factors, such as the complexity of the password (e.g. a mix of letters, numbers, and special characters), the strength of the password hashing algorithm used to store the password, and the effectiveness of the password management system in which the password is stored.

Therefore, it is not correct to say that 12-length passwords are always safe. **It is important to use strong, complex passwords and to store them securely in a password management system in order to maximize their security**.

These lines define a list of binary strings, each of which represents an ASCII character from the string `ascii_characters`.

The `ascii_characters` string is constructed by concatenating the `string.ascii_letters`, `string.digits`, and `string.punctuation strings`, which contain all the ASCII letters, digits, and punctuation marks, except the single and double quotation marks, the square brackets, the backslash, and the backtick.

The `allowed_characters` list is constructed using a list comprehension, which is a compact and efficient way to create a list in Python. The list comprehension iterates over the characters in the `ascii_characters` string, and for each character it applies the following steps:

1. It calls the `ord` function, which returns the ASCII code of the character as an integer.
2. It converts the ASCII code to a binary string using the `bin` function.
3. It strips the "0b" prefix from the binary string using string slicing.
4. If the resulting binary string is not 8 bits long, it pads the string with zeros on the left side until it is 8 bits long.
5. It appends the resulting binary string to the `allowed_characters` list.

```python
allowed_special_characters = "".join( i for i in string.punctuation if i not in "\'\"[\\]`")
ascii_characters  = string.ascii_letters + string.digits + allowed_special_characters
allowed_characters = [bin(ord(i))[2:].zfill(8) for i in ascii_characters]
```

This code defines a `PasswordGenerator` class that can be used to generate random passwords. The `PasswordGenerator` class has a single method, `get_char`, which generates a random bit string using the `magic_randomizer` function and returns it.

The `PasswordGenerator` class has an instance variable, `result_collection`, which is a list that is used to store the random bit strings generated by the `magic_randomizer` function. The `get_char` method first checks to see if the `result_collection` list is empty or has less than 5 elements. If so, it calls the `magic_randomizer` function to generate random bit strings and adds them to the `result_collection` list.

Once the result_collection list has at least 5 elements, the `get_char` method pops an element from the list and returns it. This ensures that the `get_char` method always returns a random bit string from the `result_collection` list, and refills the list with new random bit strings if it runs out.

The code also creates an instance of the PasswordGenerator class and assigns it to the password_gen_ob variable. This instance can be used to generate random passwords.

```python
class PasswordGenerator:
    def __init__(self):
        self.result_collection = list()
    def get_char(self):
        if len(self.result_collection) < 5:
            self.result_collection += magic_randomizer()
        return self.result_collection.pop()
password_gen_ob = PasswordGenerator()
```

This code constructs a password by concatenating the characters that correspond to the bit strings in the counts dictionary. The counts' list contains the measurement results of a quantum circuit that generates random bit strings, and the `allowed_characters` list contains the binary representations of the ASCII characters that are allowed in the password.

The code uses a for loop to iterate over the measurements in the list, which are the bit strings observed as measurement outcomes. For each bit string, the code does the following:

- If the length of the password is greater than 12, the loop breaks and the remaining bit strings are not processed. This is done to limit the length of the password.
- If the bit string is in the `allowed_characters` list, the code checks whether it represents an ASCII letter, a digit, or a punctuation mark. This is done using the in operator and the `string.ascii_letters`, `string.digits`, and `string.punctuation` strings, which contain all the ASCII letters, digits, and punctuation marks, respectively.
- If the bit string represents an ASCII letter, the code increments the `chars_in_a_row` variable by 1. This variable keeps track of the number of consecutive ASCII letters in the password. If this number is greater than 3, the code continues to the next iteration of the loop without adding the character to the password. This is done to ensure that the password does not contain too many consecutive letters.
- If the bit string represents a digit, the code increments the `numbers_in_a_row` variable by 1. This variable keeps track of the number of consecutive digits in the password. If this number is greater than 3, the code continues to the next iteration of the loop without adding the character to the password. This is done to ensure that the password does not contain too many consecutive digits.
- If the bit string represents a punctuation mark, the code increments the `punc_in_a_row` variable by 1. This variable keeps track of the number of consecutive punctuation characters in the password. If this number is greater than 3, the code continues to the next iteration of the loop without adding the character to the password. This is done to ensure that the password does not contain too many consecutive punctuation marks.

```python
def generate_safe_password(len_password=12):
    """
    A password is considered strong if:
    12 characters length or more
    1 digit or more
    1 symbol or more
    1 uppercase letter or more
    1 lowercase letter or more
    """
    password = ""
    chars_in_a_row = 0
    numbers_in_a_row = 0
    punc_in_a_row = 0
    while True:
        if len(password) == len_password:
            return password
        char = password_gen_ob.get_char()
        if len(password) == 11:
            if not any([i in allowed_special_characters for i in password]):
                if char in allowed_special_characters:
                    password += chr(int(char, 2))
                    continue
                else:
                    continue
        if char in allowed_characters:
            if char in string.ascii_letters:
                chars_in_a_row += 1
                numbers_in_a_row = punc_in_a_row = 0
                if chars_in_a_row > 3:
                    continue
            if char in string.digits:
                numbers_in_a_row += 1
                chars_in_a_row = punc_in_a_row = 0
                if numbers_in_a_row > 3:
                    continue
            if char in allowed_special_characters:
                punc_in_a_row += 1
                chars_in_a_row = numbers_in_a_row = 0
                if punc_in_a_row > 3:
                    continue
            password += chr(int(char, 2))

password = generate_safe_password()
print("Secure password generated: ", password)
```

```python
# Test
print("Password test: ", password)
class TestPassword(unittest.TestCase):
    def test_password_strength(self):
        self.assertTrue(len(password) == 12)
        self.assertTrue(len(re.compile("[a-z]").findall(password)) > 0)
        self.assertTrue(len(re.compile("[A-Z]").findall(password)) > 0)
        self.assertTrue(len(re.compile(f"[{allowed_special_characters}]").findall(password)) > 0)
unittest.main(argv=[''], verbosity=2, exit=False)
```

### Dice number generator


A dice number generator is a tool or program that generates random numbers that simulate the roll of a die. Many dice number generators use algorithms or other methods to generate random numbers that mimic the probability distribution of a fair die roll. The DiceRoll class is a simple class that uses the `magic_randomizer` function to generate a list of random bit strings, and provides a `get_result` method that returns a single random bit string from the list. The class maintains a list of random bit strings in the `result_collection` attribute, which is initialized to an empty list.

The get_result method first checks if the `result_collection` is empty, and if it is, it calls the `magic_randomizer` function to generate a new list of random bit strings. The method then removes and returns the last element from the `result_collection` list.

```python
class DiceRoll:
    def __init__(self):
        self.result_collection = list()
    def get_result(self):
        if len(self.result_collection) == 0:
            self.result_collection = magic_randomizer()
        return self.result_collection.pop()
dice_roll_ob = DiceRoll()
```

The `dice_roll` function is a function that uses the `magic_randomizer` function and the `DiceRoll` class to generate random bit strings and return a random integer in the range 1-6, which can be used to simulate a dice roll. The function creates a `DiceRoll` object and uses the `get_result` method to generate a random bit string. It then converts the bit string to an integer using the `int` function, and applies the modulo 7 operation to the integer to ensure that it is in the range 1-6. The resulting integer is then returned.

```python
def dice_roll():
    while True:
        roll_res = int(dice_roll_ob.get_result(), 2)%7
        if 0 < roll_res < 7:
            return  roll_res
print("El resultado del lanzamiento es: ", dice_roll())
```

This test ensures that the `dice_roll` function is returning numbers with the expected distribution.

```python
# Test
results = [dice_roll() for _ in range(100)]
class TestDiceRoll(unittest.TestCase):
    def test_valid_results(self):
        self.assertTrue(all(x in [1,2,3,4,5,6] for x in results))
    def test_dice_roll(self):
        # Test that the function returns numbers with the expected distribution
        self.assertTrue(abs(sum(results) / len(results) - 3.5) < 0.5)
unittest.main(argv=[''], verbosity=2, exit=False)
```

## Random hexadecimal value generator

Random hexadecimal values are often used as a way to generate random numbers or strings in computer programs. Because hexadecimal is a base-16 numbering system, each digit in a hexadecimal number can represent 16 different values, which allows for a more compact representation of random numbers compared to other numbering systems. This can be particularly useful for generating unique identifiers or keys, such as for objects in a database or for encrypting data. Hexadecimal values can also be more easily converted to and from other numbering systems, such as binary and decimal, which can make them more versatile for a variety of purposes.

The `HexadecimalGenerator` class is a simple class that uses the `magic_randomizer` function to generate a list of random bit strings, and provides a `get_hexa` method that returns a single random bit string from the list. The class maintains a list of random bit strings in the `result_collection` attribute, which is initialized to an empty list.

The `get_hexa` method first checks if the `result_collection` is empty, and if it is, it calls the `magic_randomizer` function to generate a new list of random bit strings. The method then removes and returns the last element from the `result_collection` list.

```python
class HexadecimalGenerator:
    def __init__(self):
        self.result_collection = list()
    def get_hexa(self):
        if len(self.result_collection) == 0:
            self.result_collection = magic_randomizer()
        return self.result_collection.pop()
hexa_gen_ob = HexadecimalGenerator()
```

The `generate_hexadecimal` function is a simple function that uses the `HexadecimalGenerator` class to generate a random hexadecimal string of a specified length. The function takes a `hex_string_len` parameter that specifies the length of the hexadecimal string to be generated. It creates a `HexadecimalGenerator` object and uses the `get_hexa` method to generate random bit strings. For each bit string, it converts the bit string to a hexadecimal string using the `hex` function, and appends the resulting hexadecimal string to the `hex_string` variable. The final `hex_string` is then returned.

```python
def generate_hexadecimal(hex_string_len=100):
    hex_string = ""
    for _ in range(hex_string_len):
        hex_string += hex(int(hexa_gen_ob.get_hexa(), 2))[2:]
    return hex_string

hexa_string = generate_hexadecimal()
print("Cadena aleatoria de caracteres en hexadecimal generada: ", hexa_string)
```

### Random number generator between two numbers

Random numbers are often used in computer programs to simulate random events or to generate data that has no recognizable pattern. In some cases, it is important to generate random numbers within a specific range, such as when generating random numbers for use in a statistical analysis or a Monte Carlo simulation. Generating random numbers within a specific range allows the program to generate numbers that are relevant to the problem at hand and are more likely to be useful for the intended purpose.

This code defines a class called `RandomNumberGenerator`. The class has a constructor method that initializes an empty list called `result_collection`, and a get_number method that returns a random number from `result_collection`. If `result_collection` is empty, the `get_number` method calls the `magic_randomizer` function to generate a new set of random numbers, which are then stored in `result_collection`.

The final line of code creates an instance of the `RandomNumberGenerator` class and stores it in the `rand_num_ob` variable. This object can be used to generate random numbers using the `get_number` method.

```python
class RandomNumberGenerator:
    def __init__(self):
        self.result_collection = list()
    def get_number(self):
        if len(self.result_collection) == 0:
            self.result_collection = magic_randomizer()
        return self.result_collection.pop()
rand_num_ob = RandomNumberGenerator()
```

This code defines a function called `get_random_number` that generates a random number within a specified range. The function takes two optional arguments: `min_n`, which specifies the minimum value of the range, and `max_n`, which specifies the maximum value of the range.

The function first checks that `min_n` is less than `max_n`. If `min_n` is greater than or equal to `max_n`, a `ValueError` is raised.

Next, the function calls the `get_number` method of the `rand_num_ob` object to generate a random number. This number is then used to index into a list of numbers within the specified range, and the indexed value is returned as the result. This allows the function to return a random number within the specified range

```python
def get_random_number(min_n=0, max_n=100):
    if min_n >= max_n:
        raise ValueError("El primer valor debe ser menor que el segundo valor")
    num = rand_num_ob.get_number()
    range_nums = [i for i in range(min_n, max_n)]
    return  range_nums[int(num, 2)%len(range_nums)]

randon_num = get_random_number(-10, 3)
print(f"Randon number generated: ", randon_num)
```

```python
# Test
class TestRandomNumberGenerator(unittest.TestCase):
    def test_get_random_number(self):
        # Test that get_random_number raises a ValueError if min_n >= max_n
        with self.assertRaises(ValueError):
            get_random_number(min_n=10, max_n=10)

        # Test that get_random_number returns a number within the specified range
        self.assertGreaterEqual(get_random_number(min_n=-150, max_n=0), -150)
        self.assertLess(get_random_number(min_n=0, max_n=1000), 1000)
        self.assertGreaterEqual(get_random_number(min_n=1, max_n=5), 1)
        self.assertLess(get_random_number(min_n=151, max_n=168), 168)

unittest.main(argv=[''], verbosity=2, exit=False)
```
