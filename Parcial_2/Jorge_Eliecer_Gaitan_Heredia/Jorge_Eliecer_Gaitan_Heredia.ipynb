```python
# punto 1
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute,IBMQ
q = QuantumRegister(6,'q')
s = QuantumRegister(1, 'scratch')
c = ClassicalRegister(6,'c')

cir = QuantumCircuit(q,c,s)
cir.h(q)
cir.x(6)
cir.h(6)
cir.cx(0,6)
cir.cx(1,6)
cir.cx(2,6)
cir.cx(3,6)
cir.cx(4,6)
cir.cx(5,6)
cir.barrier()
cir.h(q[0:6])
cir.measure(q, c)
cir.draw()
```
