```python
# Punto 1.
# A continuación se desarrolla el primer punto del parcial

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute,IBMQ
from qiskit.tools.monitor import job_monitor

q = QuantumRegister(6,'q')
s = QuantumRegister(1, 'scratch')
c = ClassicalRegister(6,'c')

circuit = QuantumCircuit(q, c, s)

circuit.x(s[0])
circuit.h(s[0])
circuit.h(q[0:6])
circuit.cx(q[0:6], s[0])
circuit.barrier(q)
circuit.h(q[0:6])
circuit.measure(q, c)

circuit.draw()
```
