```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from qiskit import QuantumRegister,ClassicalRegister
```

Pregunta 1 (60 puntos)

```python
#se defina el circuito resgistros
q = QuantumRegister(6,'q')
c = ClassicalRegister(6,'c')
scratch_line = QuantumRegister(1, 'scratch')
#se agregan al circuito
qcQuestion1 = QuantumCircuit(q, scratch_line, c)
#se definen los elementos del circuito
qcQuestion1.h(q)
qcQuestion1.x(6)
qcQuestion1.h(6)
qcQuestion1.cx(q,6)
qcQuestion1.barrier()
qcQuestion1.h(0)
qcQuestion1.h(1)
qcQuestion1.h(2)
qcQuestion1.h(3)
qcQuestion1.h(4)
qcQuestion1.h(5)
qcQuestion1.measure(q,c)

#se pinta
qcQuestion1.draw(initial_state=True)
```
