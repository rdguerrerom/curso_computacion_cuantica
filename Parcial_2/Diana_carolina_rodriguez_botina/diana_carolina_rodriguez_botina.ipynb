# Parcial Cuantica

Seccion para la ejecucion de librerias en general, no hay necesidad de declarar los mismos import para los 2 circuitos

```python
!pip install git+https://github.com/qiskit-community/qiskit-textbook.git#subdirectory=qiskit-textbook-src
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
```


#  **Primer punto**



```python
q = QuantumRegister(6,'q')
s = QuantumRegister(1,'scratch')
c=ClassicalRegister(6,'c')


qcp1 = QuantumCircuit(q,s,c)
qcp1.h(q)
qcp1.x(s)
qcp1.h(s)

qcp1.cx(0,6)
qcp1.cx(1,6)
qcp1.cx(2,6)
qcp1.cx(3,6)
qcp1.cx(4,6)
qcp1.cx(5,6)

qcp1.barrier(q)

qcp1.h(q)

qcp1.measure(0,0)
qcp1.measure(1,1)
qcp1.measure(2,2)
qcp1.measure(3,3)
qcp1.measure(4,4)
qcp1.measure(5,5)

qcp1.draw()
```

# **Segundo Punto**

```python
q = QuantumRegister(16,'q')
c = ClassicalRegister(16,'meas')

qcp2 = QuantumCircuit(q,c)

qcp2.h(q)

qcp2.barrier(q)

qcp2.measure(q,c)

qcp2.draw()
```
