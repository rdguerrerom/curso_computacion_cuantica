```python
!pip install git+https://github.com/qiskit-community/qiskit-textbook.git#subdirectory=qiskit-textbook-src
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

```python
qr = QuantumRegister(6, 'q')
anc = QuantumRegister(1, 'scracth')
cr = ClassicalRegister(6, 'c')
qc = QuantumCircuit(qr, anc, cr)

qc.x(anc[0])
qc.h(anc[0])
qc.h(qr[0:6])
qc.cx(qr[0:6], anc[0])
qc.barrier(qr)
qc.measure(qr, cr)

qc.draw()
```

```python
qr_2 = QuantumRegister(16, 'q')
cr_2 = ClassicalRegister(16, 'c')
qc_2 = QuantumCircuit(qr_2,cr_2)

qc_2.h(qr_2[0:16])
qc_2.barrier(qr_2)
qc_2.measure(qr_2, cr_2)

qc_2.draw()
```
