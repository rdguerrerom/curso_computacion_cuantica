```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
```

```python
from qiskit import QuantumRegister, ClassicalRegister
```

```python
#Punto 1 Parcial
qc_ha = QuantumRegister(6, 'q')
qc_hs = QuantumRegister(1, 'scratch')
qc_hc = ClassicalRegister(6, 'c')

qc = QuantumCircuit(qc_ha, qc_hs, qc_hc)

qc.x(qc_hs[0])
qc.h(qc_hs[0])
qc.h(qc_ha[0:6])
qc.cx(qc_ha[0:6], qc_hs[0])
qc.barrier(qc_ha)
qc.h(qc_ha[0:6])
qc.measure(qc_ha, qc_hc)

qc.draw()
```
